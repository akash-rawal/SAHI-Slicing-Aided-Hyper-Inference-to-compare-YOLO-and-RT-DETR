```python
import json

with open("instances_gt.json", "r") as f:
    data = json.load(f)


id_map = {}
for i, img in enumerate(data["images"], start=1):
    id_map[img["id"]] = i
    img["id"] = i


for ann in data["annotations"]:
    ann["image_id"] = id_map[ann["image_id"]]

with open("instances_gt_fixed.json", "w") as f:
    json.dump(data, f)

print("Fixed COCO file saved as instances_gt_fixed.json")
```
